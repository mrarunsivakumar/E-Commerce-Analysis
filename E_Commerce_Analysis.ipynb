```python
# Function to extract specific detail from the list of dictionaries
def extract_detail(detail_list, key):
    for detail_dict in detail_list:
        if key in detail_dict:
            return detail_dict[key]
    return None

# Extracting specific details
df['style_code'] =df['product_details'].apply(lambda x: extract_detail(x, 'Style Code'))
df['closure'] = df['product_details'].apply(lambda x: extract_detail(x, 'Closure'))
df['pockets'] = df['product_details'].apply(lambda x: extract_detail(x, 'Pockets'))
df['fabric'] = df['product_details'].apply(lambda x: extract_detail(x, 'Fabric'))
df['pattern'] = df['product_details'].apply(lambda x: extract_detail(x, 'Pattern'))
df['color'] = df['product_details'].apply(lambda x: extract_detail(x, 'Color'))

# Drop the original 'product_details' column if needed
df.drop(columns=['product_details'], inplace=True)
```

```python
categorical_columns = ['style_code', 'fabric', 'pattern', 'color','closure','pockets']
for column in categorical_columns:
    df[column].fillna(df[column].mode()[0], inplace=True)
```

```python
# Find the most frequent brand name excluding 'Unknown'
most_frequent_brand = df[df['seller'] != 'Unknown']['seller'].mode()[0]

# Replace 'Unknown' brand values with the most frequent brand name
df.loc[df['seller'] == 'Unknown', 'seller'] = most_frequent_brand
```

```python
import re

# Function to clean seller names
def clean_seller(seller):
    # Regular expression pattern to extract seller names
    pattern = r'^([^0-9(]+)'  # Match anything that is not a digit or '('

    # Extract seller name using regular expression
    match = re.match(pattern, seller)
    if match:
        return match.group(1).strip()  # Return the matched group without leading/trailing spaces
    else:
        return seller  # Return original seller if no match found

# Apply the clean_seller function to the 'seller' column
df['seller'] = df['seller'].apply(clean_seller)

# Display the cleaned seller names
print(df['seller'].unique())
```

```python
# Remove " off" suffix and "%" symbol from the 'discount' column
df['discount'] = df['discount'].str.replace(' off', '').str.replace('%', '')

# Convert the 'discount' column to numeric format
df['discount'] = pd.to_numeric(df['discount'])
```

```python
# Remove commas from 'actual_price' column
df['actual_price'] = df['actual_price'].str.replace(',', '')

# Convert 'actual_price' column to numeric
df['actual_price'] = pd.to_numeric(df['actual_price'])
```

```python
# Convert 'average_rating' column to float
df['average_rating'] = pd.to_numeric(df['average_rating'], errors='coerce')
```

```python
# Remove commas from selling_price column
df['selling_price'] = df['selling_price'].str.replace(',', '')

# Convert selling_price column to numeric format
df['selling_price'] = pd.to_numeric(df['selling_price'])
```

```python
df.isnull().sum()
```

```python
# Replace null values in 'actual_price', 'average_rating', 'discount', and 'selling_price' columns
df['actual_price'].fillna(0, inplace=True)  # Replace null values with 0
df['average_rating'].fillna(0, inplace=True)  # Replace null values with 0
df['discount'].fillna(0, inplace=True)  # Replace null values with 0
df['selling_price'].fillna(df['selling_price'].mean(), inplace=True)  # Replace null values with the mean of 'selling_price'
```

```python
# Calculate the difference between actual price and selling price
df['price_difference'] = df['actual_price'] - df['selling_price']

# Group by brand and calculate the average difference
grouped_df = df.groupby('brand')['price_difference'].mean().reset_index()

# Sort by the average difference
grouped_df = grouped_df.sort_values(by='price_difference', ascending=False)

# Select top 10 brands
top_10_brands = grouped_df.head(10)

# Plotting the bar chart
plt.figure(figsize=(12, 8))
bars = plt.bar(top_10_brands['brand'], top_10_brands['price_difference'], color='skyblue')

# Add label values on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), va='bottom', ha='center')

plt.xlabel('Brand')
plt.ylabel('Average Price Difference')
plt.title('Top 10 Brands: Average Difference between Actual Price and Selling Price')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Calculate the difference between actual price and selling price
df['price_difference'] = df['actual_price'] - df['selling_price']

# Group by seller and calculate the average difference
grouped_df = df.groupby('seller')['price_difference'].mean().reset_index()

# Sort by the average difference
grouped_df = grouped_df.sort_values(by='price_difference', ascending=False)

# Select top 10 sellers
top_10_sellers = grouped_df.head(10)

# Plotting the bar chart
plt.figure(figsize=(12, 8))
bars = plt.bar(top_10_sellers['seller'], top_10_sellers['price_difference'], color='skyblue')

# Add label values on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), va='bottom', ha='center')

plt.xlabel('Seller')
plt.ylabel('Average Price Difference')
plt.title('Top 10 Sellers: Average Difference between Actual Price and Selling Price')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
